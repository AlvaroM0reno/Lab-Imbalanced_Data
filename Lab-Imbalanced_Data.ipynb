2. Intentaremos predecir la variable Churn utilizando una regresión logística sobre las variables tenure, Senior_Citizen, Monthly_Charges.

```python
# Paso 1: Importar las bibliotecas necesarias
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Paso 3: Preparar los datos, Seleccionamos las columnas relevantes y preprocesamos los datos según sea necesario.
# Convertir la variable Churn a binaria
df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)

# Seleccionar las columnas necesarias
X = df[['tenure', 'SeniorCitizen', 'MonthlyCharges']]
y = df['Churn']
```

```python
# Paso 4: Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Paso 5: Entrenar el modelo de regresión logística.
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Paso 6: Realizar predicciones y evaluar el modelo.
# Predicciones
y_pred = model.predict(X_test)

# Evaluar el modelo
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```
